# Import Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('../data/train.csv')
```

# Data Cleaning

```python
df.info()
df.shape
```

```python
# check for duplicates
df.duplicated().value_counts()
```

```python
# check for missing value
df.isna().sum().sort_values(ascending = False).head(10)
```

At this point, decided features to work with:

Feature|Type|Description
---|---|---
flat_type|Categorical| type of the resale flat unit, e.g. 3 ROOM 
pri_sch_name|Categorical|name of the nearest primary school
sec_sch_name|Categorical|name of the nearest secondary school
bus_interchange|Binary| boolean value if the nearest MRT station is also a bus interchange
mrt_interchange|Binary| boolean value if the nearest MRT station is a train interchange station
market_hawker|Binary| boolean value if resale flat has a market or hawker centre in the same block
multistorey_carpark|Binary|boolean value if resale flat has a multistorey carpark in the same block
year_completed|Int|year which construction was completed for resale flat
Tranc_Year|Int|year of resale transaction
floor_area_sqft|Float|floor area of the resale flat unit in square feet
pri_sch_nearest_distance|Float|distance (in metres) to the nearest primary school
sec_sch_nearest_dist|Float|distance (in metres) to the nearest secondary school

```python
#features
categories = ['flat_type','pri_sch_name','sec_sch_name'] 
binaries = ['bus_interchange','mrt_interchange','market_hawker','multistorey_carpark'] 
numerics = ['year_completed','Tranc_Year','floor_area_sqft','pri_sch_nearest_distance','sec_sch_nearest_dist']
#target
target = ['resale_price']

hdb = df[categories].join(df[binaries]).join(df[numerics]).join(df[target])
```

```python
hdb.head()
hdb.info()
```

# Feature engineering

Transform 'market_hawker' and 'multistorey_carpark' columns to binary

```python
hdb ['market_hawker'] = (hdb ['market_hawker'] == 'Y').astype(int) # 1 = available, 0 = not available
hdb ['multistorey_carpark'] = (hdb ['market_hawker'] == 'Y').astype(int) # 1 = available, 0 = not available
```

```python
hdb [binaries].head()
```

Create interaction item: 
1. both_bus_mrt_interchange --> the instance where both bus and mrt interchange is available
2. flat_age_when_purchase --> flat age at the time of purchase

```python
hdb ['both_bus_mrt_interchange'] = hdb ['bus_interchange'] * hdb ['mrt_interchange'] # =1 only if both are available
binaries.append('both_bus_mrt_interchange')
hdb ['flat_age_when_purchase'] = hdb ['Tranc_Year'] - hdb ['year_completed'] # flat age at the time of purchase
numerics.append('flat_age_when_purchase')
```

# EDA

```python
sns.pairplot(hdb, y_vars=target, x_vars=numerics)
```

```python
hdb.hist(figsize=(10, 10));
```

> * Initial observation from various features (vs target) shows that resale price have similar distribution with flat age and floor area
> * Two features: market_hawker and multistorey_carpark do not have sufficient data on *'available'* values, may want to consider to drop in second model iteration. 

```python
plt.figure(figsize=(10,10))
sns.heatmap(hdb.corr(), annot=True);
```

> * Assessing different features against **resale price**, `floor_area_sqft` (0.65) and `year_completed` (0.35) are among the highest correlation coefficient, with larger floor area and latest completed year having higher resale price.
> * Meanwhile, there is a potential **time leakage** phenomenon as seen from correlation between `year_completed` and `bus_interchange`,`mrt_interchange`,`floor_area_sqft`.
> * To consider train-test split based on time series.

# Model Prep

Recall different features:
* categories
* numerics
* binaries

> Categories will be encoded, while numerics will be scaled. Afterwards, three of them will be joined together.

## Train-Test Split

```python
X = hdb.drop(columns = 'resale_price')
y = hdb['resale_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split (X,y,test_size = 0.5, random_state = 99) # initiate with 50/50 train-test ratio
```

## Categories: One Hot Encoder

```python
X_train_cat = X_train[categories]
```

```python
X_test_cat = X_test[categories]
```

```python
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore',sparse = False)
```

```python
# fit and transform
X_train_cat_enc = enc.fit_transform(X_train_cat)
X_test_cat_enc = enc.transform(X_test_cat)
```

```python
# move into new dataframe
X_train_cat_enc_df = pd.DataFrame(X_train_cat_enc, columns = enc.get_feature_names_out(), index = X_train_cat.index) # input index to keep original indices
X_test_cat_enc_df = pd.DataFrame(X_test_cat_enc, columns = enc.get_feature_names_out(), index = X_test_cat.index)
```

```python
X_train_cat_enc_df.head()
```

```python
#fit_transform
X_train_cat_pri = enc.fit_transform(X_train_cat[['pri_sch_name']])
X_test_cat_pri = enc.transform(X_test_cat[['pri_sch_name']])
```

```python
X_train_cat[enc.categories_[0]] = X_train_cat_pri.toarray()
```

```python
X_test_cat[enc.categories_[0]] = X_test_cat_pri.toarray()
```

## Numerics: Standard Scaler

```python
X_train_num = X_train[numerics]
```

```python
X_test_num = X_test[numerics]
```

```python
ss = StandardScaler()
```

```python
X_train_num_scaled = ss.fit_transform (X_train_num)
X_test_num_scaled = ss.transform (X_test_num)
```

```python
X_train_num_scaled_df = pd.DataFrame(data = X_train_num_scaled, index = X_train_num.index, columns = X_train_num.columns) # input index to keep original indices
X_train_num_scaled_df
```

```python
X_test_num_scaled_df = pd.DataFrame(data = X_test_num_scaled, index = X_test_num.index, columns = X_test[numerics].columns)
X_test_num_scaled_df
```

## Binaries: No Action Needed

```python
X_train_bin = X_train[binaries]
X_test_bin = X_test[binaries]
```

## Combine all 3 train/test datasets

```python
X_train_prep = X_train_cat_enc_df.join([X_train_num_scaled_df,X_train_bin])
X_test_prep = X_test_cat_enc_df.join([X_test_num_scaled_df,X_test_bin])
```

```python
# Validate
X_train_prep
```

```python
X_test_prep
```

# Modeling

```python
# instantiate 3 different models
lr = LinearRegression()
ridge = RidgeCV(alphas=np.linspace(.1, 10, 100))
#lasso = LassoCV(n_alphas=200)
```

```python
lasso = LassoCV(n_alphas=500)
```

```python
# cross-validate model scores
lr_scores = cross_val_score(lr, X_train_prep, y_train, cv=3)
print(f'linear reg score = {lr_scores.mean()}')

ridge_scores = cross_val_score(ridge, X_train_prep, y_train, cv=3)
print(f'ridge score = {ridge_scores.mean()}')

lasso_scores = cross_val_score(lasso, X_train_prep, y_train, cv=3)
print(f'lasso score = {lasso_scores.mean()}')
```

> Linear and Ridge seem to have similar and joint highest score. We will try to fit these two models in the next section.

## Model 1 : Linear Regression

```python
# fit
lr.fit(X_train_prep, y_train)
```

```python
# train-test score
print(lr.score (X_train_prep, y_train))
print(lr.score (X_test_prep, y_test))
```

```python
# predict y hat
y_pred_train_lr = lr.predict (X_train_prep)
y_pred_test_lr = lr.predict (X_test_prep)
```

```python
sns.jointplot(x = y_pred_train_lr, y= y_train);
```

```python
sns.jointplot(x = y_pred_test_lr, y= y_test);
```

## Model 2 : Ridge Regression

```python
# fit
ridge.fit(X_train_prep, y_train)
```

```python
# train-test score
print(ridge.score (X_train_prep, y_train))
print(ridge.score (X_test_prep, y_test))
```

```python
# predict y hat
y_pred_train_ridge = ridge.predict (X_train_prep)
y_pred_test_ridge = ridge.predict (X_test_prep)
```

```python
sns.jointplot(x = y_pred_train_ridge, y= y_train);
```

```python
sns.jointplot(x = y_pred_test_ridge, y= y_test);
```
